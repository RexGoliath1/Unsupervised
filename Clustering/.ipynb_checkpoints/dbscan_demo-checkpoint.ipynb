```python
# Set up and Generate Data
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

plt.style.use('ggplot')
%matplotlib inline
```

```python
X,label = make_moons(n_samples=200, noise=0.1, random_state=19)
print(X[:5,])
```

# Train the Model

```python
# Model = DBSCAN
model = DBSCAN(eps=0.25, min_samples=12).fit(X)
print(model)
```

# Access the cluster labels and core points

```python
model.labels_
```

# Find Core Points

```python
model.core_sample_indices_
```

# Plot the cluster in feature space

```python
fig, ax = plt.subplots(figsize=[10,8])
sctr = ax.scatter(X[:,0],X[:,1], c=model.labels_, s=140, alpha=0.9,
                 cmap=plt.cm.Set1)
fig.show()
```

## Silhouette Score

```python
xy = model.fit_predict(X)

avg_sil = silhouette_score(X, model.labels_)

print("Silhouette Score %f" % avg_sil)
```
